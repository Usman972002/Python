# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
ds = pd.read_csv('BigBasket.csv')
trans = []
for i in range(0,7218):
  trans.append([str(ds.values[i,j]) for j in range(20)])
```

```python
# print(trans)
```

# Training Apriori

```python
from apyori import apriori
basket  = apriori(transactions = trans,min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

# Visualizing Raw Results

```python
results = list(basket)
results
```

# Better Visualization

```python
def inspect(results):
    product1         = [tuple(result[2][0][0])[0] for result in results]
    product2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(product1, product2, supports, confidences, lifts))
DataFrame_intelligence = pd.DataFrame(inspect(results), columns = ['product1', 'product1', 'Support', 'Confidence', 'Lift'])
```

```python
DataFrame_intelligence
```

```python
# Best Lifting Items
DataFrame_intelligence.nlargest(n=10,columns='Lift')
```

```python
# Best Support Items in Basket
DataFrame_intelligence.nlargest(n=10,columns='Support')
```

```python
# Best Confidence Items in Basket
DataFrame_intelligence.nlargest(n=10,columns='Confidence')
```
